###A common use for NumPy is for matrix algebra for linear regressions.###

For this example, I am going to show how NumPy calculates OLS, Ordinary Least Squares.


__First let's create our X and y data sets in this supervised example.__

*Recall that X is a maxtrix and y is a vector. You can see a visual example below. I also randomly generated some numbers to plot and added random noise.*

Here is an example of what X can look like:

<center>
$\begin{bmatrix} 
x_{1,1} & x_{1,2} & x_{1,3} \\ 
x_{2,1} & x_{2,2} & x_{2,3} \\ 
x_{3,1} & x_{3,2} & x_{3,3}  \end{bmatrix}$</center>

<hr noshade size=2>

Alternatively, y looks like this:
<center>
$\left[ \begin{array}{c} y_1 \\ y_2\\y_3 \end{array} \right] $</center>

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

X = np.array(zip(np.ones(100),np.random.rand(100)))
y = X[:,1] + np.random.rand(100)*.8

plt.scatter(X[:,1],y)
```

###What about Beta?###

$\beta$ is also a vector and what we are trying to solve for. 
We aim to minimize the sum of the squared residuals using the OLS equation:
<p><center><h3>
$\text{min }_\beta||y-X\beta||^2$</center></p></h3>

__How do you solve for $\beta$?__
<h4>
$||y-X\beta||^2$
<p>$ = (y-X\beta)^T(y-X\beta)$</p>
<p>$ = y^Ty -2\beta^TX^Ty + \beta^TX^TX\beta$</p>
<p>We can differentiate this function with respect to $\beta$ $ -2X^Ty + 2 X^TX\beta$.</p>
<p>The second derivative is $ 2X^TX \geq 0$. </p>
<p>Recall that we can find the minimum from setting the first derivative equal to 0 and solving for $\beta$.</p>
<p>$ -2X^Ty + 2 X^TX\beta = 0$</p>
<p>$ 2 X^TX\beta = 2X^Ty$</p>
<p>$ X^TX\beta = X^Ty$</p>
<p>$ \beta = (X^TX)^{-1}X^Ty$</p></h4>

__Tada!__

In NumPy, this is expressed below in this code: 

```python
beta = np.dot(np.linalg.inv(np.dot(X.T,X)),np.dot(X.T,y))

#Graphing the best fit line using OLS
plt.scatter(X[:,1],y)
plt.plot(X[:,1],np.dot(X,beta))
plt.ylabel('Y')
plt.xlabel('x')
plt.axis([0, 1, 0, 2])
plt.show()
```

#This is how we create a best fit line using OLS!#

For more documentation on deriving the OLS equation, this is a great read: <a href="http://isites.harvard.edu/fs/docs/icb.topic515975.files/OLSDerivation.pdf"> Harvard OLS Derivation</a>

Please note that there are other ways to run a linear regression using OLS such as using the StatsModel Package with some additional fancy methods that take care of non invertable matrices.
